```python
import data_reduce as dr
import numpy as np
```

#### Generate mock data

```python
m_points = int(1e5)
X = np.linspace(-20, 20, m_points)
Y = X*np.sin(X/2) + 2*np.cos(X) + 0*np.random.normal(0, 1, m_points)
Z = np.vstack((X, Y)).T
```

#### Reduce

```python
Z_culled = dr.VWReducer().reduce(Z, 25)
```

```python
fig = dr.show_reduction(Z, Z_culled)
fig.update_layout(width=400, height=300)
fig.show()
```
